```python
# code Edge et Digraph du CM5
class Edge(object):
    def __init__(self, src, dest):
        self.src = src
        self.dest = dest
    def getSource(self):
        return self.src;
    def getDestination(self):
        return self.dest
    def __str__(self):
        return self.src.getName + '->' + self.dest.getname()
    
class Digraph(object):
    # nodes est une liste de noeuds
    # edges est un dictionnaire mappant chaque noeud à ses enfants
    def __init__(self):
        self.nodes = []
        self.edges = {}
    def addNode(self, node):
        if node in self.nodes:
            raise ValueError('Duplicate node')
        else:
            self.nodes.append(node)
            self.edges[node] = []
    def addEdge (self, edge):
        src = edge.getSource()
        dest = edge.getDestination()
        if not(src in self.nodes and dest in self.nodes):
            raise ValueError('Node not in graph')
        self.edges[src].append(dest)
    def childrenOf(self, node):
        return self.edges[node]
    def hasNode(self, node):
        return node in self.nodes
    def __str__(self):
        result = ''
        for src in self.nodes:
            for dest in self.edges[src]:
                result = result + src.getName() + '->'\
                        + dest.getName() + '\n'
        return result[:-1]

# Exemple du TP4, TD5
nodes = listeNP
print(nodes)
g = Digraph()
for n in listeNP:
      g.addNode(n)
g.addEdge(Edge(nodes[0], nodes[1]))
g.addEdge(Edge(nodes[0], nodes[2]))
g.addEdge(Edge(nodes[0], nodes[4]))
g.addEdge(Edge(nodes[0], nodes[6]))
g.addEdge(Edge(nodes[0], nodes[7]))
g.addEdge(Edge(nodes[1], nodes[2]))
g.addEdge(Edge(nodes[1], nodes[3]))
g.addEdge(Edge(nodes[2], nodes[3]))
g.addEdge(Edge(nodes[2], nodes[4]))
g.addEdge(Edge(nodes[3], nodes[4]))
g.addEdge(Edge(nodes[3], nodes[5]))
g.addEdge(Edge(nodes[3], nodes[1]))
g.addEdge(Edge(nodes[4], nodes[5]))
g.addEdge(Edge(nodes[6], nodes[5]))
g.addEdge(Edge(nodes[7], nodes[5]))
print(g)
for i in range(8):
    print(g.nodes[i])   
```

```python
#DFS et BFS du cours, peut s'appliquer sans modification pour trouver le chemin le plus court
def printPath(path):
    result = ''
    for i in range(len(path)):
        result = result + str(path[i])
        if i != len(path) - 1:
            result = result + '->'
    return result

def DFS(graph, start, end, path, shortest):
    path = path + [start]
    print('Current DFS path', printPath(path))
    if start == end:
        return path
    for node in graph.childrenOf(start):
        if node not in path: #evite les cycles
            if shortest == None or len(path) < len(shortest):
                newPath = DFS(graph, node, end ,path, shortest)
                if newPath != None:
                    shortest = newPath
    return shortest

def search(graph, start, end):
    return DFS(graph, start, end , [], None)


sp=search(g, nodes[0], nodes[5])
print('Chemin le plus court avec DFS', printPath(sp))

def BFS(graph, start, end):
    initPath = [start]
    pathQueue = [initPath]
    while len(pathQueue) != 0:
        tmpPath = pathQueue.pop(0)
        print('Current BFS path:', printPath(tmpPath))
        lastNode = tmpPath[-1]
        if lastNode == end:
            return tmpPath
        #print('lastNode', lastNode)
        for nextNode in graph.childrenOf(lastNode):
            if nextNode not in tmpPath:
                #print('Node', nextNode)
                newPath = tmpPath + [nextNode]
                pathQueue.append(newPath)
                #print('pathQ', printPath(newPath))
    #return pathQueue
    return None
sp = BFS(g, nodes[0], nodes[5])
print('Chemin le plus court avec  BFS', printPath(sp))
```

```python
#Tous les chemins
def DFS_list(graph, start, end, path, lsp):
    path = path + [start]
    print('Current DFS path', printPath(path))
    if start == end:
        return path
    for node in graph.childrenOf(start):
        if node not in path: #evite les cycles
            #if shortest == None or len(path) < len(shortest):
            #on récupère tous les chemins
                newPath = DFS_list(graph, node, end, path, lsp)
                if newPath[-1] == end:
                     lsp.append(newPath)
                #if newPath != None:
                    #shortest = newPath
    return path

lesChemins = []
path = DFS_list(g, nodes[0], nodes[5], [], lesChemins)
print('Chemin', printPath(path))
for path in lesChemins:
    print('Chemin', printPath(path))
```
